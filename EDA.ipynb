# Exploratory Data Analysis (EDA)

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
def load_data():
    data_dir = 'input/'

    df = pd.read_csv(data_dir + 'climate_change_impact.csv')
    return df
```

```python
df = load_data()
```

```python
# Distribusi Target

plt.figure(figsize=(8,4))
sns.histplot(df['Economic_Impact_Million_USD'], kde=True)
plt.title('Distribusi Economic Impact (Million USD)')
plt.show()
```

```python
# Korelasi Fitur

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribusi fitur kategorikal

# Distribusi Region
df['Region'].value_counts().plot(kind='bar')
plt.title('Distribusi Region')
plt.show()

# Distribusi Crop_Type
df['Crop_Type'].value_counts().plot(kind='bar')
plt.title('Distribusi Crop Type')
plt.show()
```

```python
# Hubungan antara fitur dan target

# Contoh: Average Temperature vs Economic Impact
sns.scatterplot(x='Average_Temperature_C', y='Economic_Impact_Million_USD', data=df)
plt.title('Average Temperature vs Economic Impact')
plt.show()

# Contoh: Crop Yield vs Economic Impact
sns.scatterplot(x='Crop_Yield_MT_per_HA', y='Economic_Impact_Million_USD', data=df)
plt.title('Crop Yield vs Economic Impact')
plt.show()
```

```python
# Boxplot outlier

plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Average_Temperature_C', 'Total_Precipitation_mm', 'Crop_Yield_MT_per_HA', 'Economic_Impact_Million_USD']])
plt.title('Boxplot Fitur Numerik')
plt.show()
```
